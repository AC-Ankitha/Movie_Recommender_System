```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast

def fetch_director(obj):
    L = []
    try:
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                L.append(i['name'])
                break  # Exit loop after finding the director
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing data: {e}")
    return L

# Apply the function to the 'crew' column
directors = movies['crew'].apply(fetch_director)
print(directors.head())  # Check the first few results
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
 new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from nltk.stem.porter import PorterStemmer

# Initialize the PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))  # Correctly pass the word to stem method
    
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Assuming vectors is your 2D array of shape (n_samples, n_features)
similarity_matrix = cosine_similarity(vectors)

# Print the similarity matrix
print(similarity_matrix)
```

```python
sorted(list(enumerate(similarity_matrix[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    # Check if the movie exists in the dataset
    if movie not in new_df['title'].values:
        print("Movie not found in dataset.")
        return
    
    # Get the index of the movie
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    # Compute similarity scores
    distances = similarity_matrix[movie_index]
    
    # Get movie indices sorted by similarity
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Return recommended movie titles
    recommended_movies = [new_df.iloc[i[0]]['title'] for i in movies_list]
    return recommended_movies

# Example usage
recommended_titles = recommend('Batman Begins')
print("Recommended movies:", recommended_titles)
```
